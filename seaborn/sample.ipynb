```python
import seaborn
from matplotlib import pyplot
import pandas as pd
import numpy as np
```

```python
days = ["mon","tue","wed","thur","fri"]
sales = [100,120,110,130,140]
df = pd.DataFrame({"Days":days,"Sales":sales})
df
```

```python
seaborn.lineplot(data = df,x = "Days" , y = "Sales")
pyplot.xlabel('Dayes in week')#THe seaborn automaticaly xlabel and ylabel as dayes and sales
pyplot.ylabel("Sales  count")#if we want to modify we have to use pyplot.xlabel,ylabel
pyplot.show()
```

```python
students = ["Arun","Binu","Ram","Vinay","Sajay"]
maths = [85,78,92,70,88]
python = [80,74,95,68,90]
df = pd.DataFrame({"Name":students,"Maths":maths,"Python":python})
df
```

```python
seaborn.lineplot(x = students,y=python)
pyplot.show()
```

```python
seaborn.barplot(x = students,y=python,palette="coolwarm",hue = python)
pyplot.show()
```

```python
x = [1,3,5,7,9]
y = [5,7,9,11,13]
seaborn.scatterplot(x = x,y = y)
pyplot.show()
```
