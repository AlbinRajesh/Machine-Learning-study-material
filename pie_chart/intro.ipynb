```python
import pandas as pd
import numpy as np
```

```python
from matplotlib import pyplot
```

```python
activities = ["sleeping","work","travel","other"]
hours = [8,8,4,4]
pyplot.pie(hours,labels=activities)
pyplot.show()
```

```python
activities = ["sleeping","skyblue","green","cyan"]
explode = (0.1,0,0,0)
colors = ["red","skyblue","green","cyan"]
pyplot.pie(hours,labels=activities,autopct = '%1.1f%%',startangle=90,colors=colors,explode=explode,shadow=True)
pyplot.show()
```

```python
categories = ["flood", "rent", "fuel", "shopping", "other"]
expenses = [500, 2000, 800, 1500, 300]   
colors = ["red", "skyblue", "green", "cyan", "orange"]
explode = [0.1, 0, 0, 0, 0]     
pyplot.pie(expenses,labels=categories,autopct='%1.1f%%',startangle=90,colors=colors,explode=explode,shadow=True)
```
