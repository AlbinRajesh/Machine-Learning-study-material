```python
ckustering
----------------
grouping the given data is called the clustering
eg:["cat","orange","rat","apple","mango"]
animals(cluster 1)> ["cat","rat"]

fruits(cluster 2)> ["orange","apple","mango"]
```

```python
K - means clustering
--------------------------------
The first step is to define k (which is number of clusters)
k = 2, cluster=2
if k = 4, cluster = 4
```

```python
Centroid 
----------
the center point of a cluster 
center of a group formed by averaging all points inside it
```

```python
the point which we define to divide the cluster is known as elbow point
eg)
customers which buy for more than 10000rs and less than 10000rs is another cluster 
here 10000 rs is the elbow point (-----the elbow point is not an element or data it is a condition------)
```

```python
the  distance of an elament from the Centroid is known as varience
we can say a cluster is a good when the varience is less among the clusters
```
