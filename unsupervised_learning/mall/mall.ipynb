```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder 
```

```python
df = pd.read_csv("Mall_Customers.csv")
df.isna().sum()
```

```python
df
```

```python
le = LabelEncoder()
df["Gender"] = le.fit_transform(df["Gender"])
```

```python
df
```

```python
x = df[["Annual Income (k$)","Spending Score (1-100)"]]
```

```python
x
```

```python

from matplotlib import pyplot as plt
```

```python
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
wcss = []#Within cluster sum of squares
for i in range(1,11):
    KMeans_new = KMeans(n_clusters=i,init = "k-means++")
    KMeans_new.fit(x)
    wcss.append(KMeans_new.inertia_)
#This code is for just the ploting below-- 
```

```python
plt.plot(range(1,11),wcss,marker = "o")
plt.show()
```

```python
KMeans_new = KMeans(n_clusters=2,init='k-means++')
KMeans_new.fit(x)
y_pred = KMeans_new.predict(x)
y_pred
```

```python
plt.scatter(x.iloc[:,0],x.iloc[:,1],c=y_pred)
plt.scatter(KMeans_new.cluster_centers_[:,0],KMeans_new.cluster_centers_[:,1])
plt.show()
```
