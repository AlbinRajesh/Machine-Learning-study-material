"In linear regression it can predict a result using one input feature"


```python
area = [2600,2800,3100,3400,3800]
price = [5600000,5800000,6100000,6600000,7200000]
```

```python
from matplotlib import pyplot

pyplot.scatter(area, price)
pyplot.xlabel("area")
pyplot.ylabel("price")
pyplot.title("Area_HousePrice")
pyplot.show()
```
