```python
# 1. Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
char_col = df.select_dtypes(include=['object']).columns
char_col
```

```python

from sklearn.preprocessing import OneHotEncoder
# 2. Create encoder
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
# 3. Fit ONLY on training data
encoder.fit(x_train[char_col])
# 4. Transform both sets using same encoder
X_train_encoded = encoder.transform(x_train[char_col])
X_test_encoded  = encoder.transform(x_test[char_col])




```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(X_train_encoded)
x_test = scaler.transform(X_test_encoded)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)  
```

```python
Y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
print("Mean Absolute Error:", mean_absolute_error(y_test, Y_pred))
```
