```python
import requests
import pandas as pd
import time

api_key = "AIzaSyD0tgyb28aXM6x5RpGmGvj1rG41kygzvYc"
queries = [
    "Data Science", "Machine Learning", "Python", "Artificial Intelligence",
    "Deep Learning", "Statistics", "Computer Vision", "Web Development",
    "Programming", "Big Data","Romance", "Thriller", "Mystery", "Fantasy", "Horror", "Science Fiction"
]

books = []

# Fetch multiple pages (up to 200 results)
for start in range(0, 200, 40):  # 0, 40, 80, 120, 160
    url = f"https://www.googleapis.com/books/v1/volumes?q={query}&startIndex={start}&maxResults=40&key={api_key}"
    response = requests.get(url)
    data = response.json()

    if "items" not in data:
        break  # stop if no more results

    for item in data["items"]:
        volume = item["volumeInfo"]
        books.append({
            "Title": volume.get("title"),
            "Authors": ", ".join(volume.get("authors", [])),
            "Publisher": volume.get("publisher"),
            "PublishedDate": volume.get("publishedDate"),
            "Category": ", ".join(volume.get("categories", [])) if "categories" in volume else None,
            "AverageRating": volume.get("averageRating"),
            "PageCount": volume.get("pageCount"),
            "Language": volume.get("language")
        })

    print(f"Fetched {len(books)} books so far...")
    time.sleep(1)  # small delay to avoid rate limiting

df = pd.DataFrame(books)
df.to_csv("books_dataset.csv", index=False)
print(f"✅ Done! Collected {len(df)} books and saved to books_dataset.csv")
```
