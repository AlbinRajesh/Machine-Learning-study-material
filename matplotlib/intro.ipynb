------Matplotlib--------
which is a python data visualization library
helps to visualize data through graph,plot and chart

eg : line,bar,pie,charts
customize the colour,linestyle etc

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot
```

```python
x = [2,3,4,5]
y = [8,9,10,11]
# x = np.array([1,2,3,4])
# y = np.array([8,9,10,11])#this another way of describe the data ,there are so many other ways
pyplot.plot(x,y)
pyplot.show()
```

```python
x = [3,2,4,5,6,7]
y = [1,4,2,5,6,7]
#this type of graph is known as marked
pyplot.plot(x,y,"*")
pyplot.show()
```

```python
x = [4,5,6,7]
y = [3,7,5,4]
pyplot.plot(x,y,marker="o")#this marker method returns a graph with o and lineb
pyplot.show()
```

```python
x = [4,5,6,7]
y = [3,7,5,4]
pyplot.plot(x,y,"o:g") #point(o=this mentions the point )   dotted line (:=this mentions the dotted line)  color(g=this mentions the color)
```

```python
# Marker properties
# ------------------------
#we can customize the marker size,marker edge color,marker face colour
x = [4,5,6,7]
y = [3,7,5,4]
pyplot.plot(x,y,marker="o",ms = 20,mfc = "r",mec = "g")#marker type = o,marker size(20), marker face colour(red),marker edge colour(green)
pyplot.show()
```

```python
x = np.array([2,4,6,8])
y = np.array([6,8,9,10])
pyplot.plot(x,y,marker = "o",ms = 25,mec="r",mfc="g")
```

```python
#multiple lines
# ====================
x1 = [1,2,3,4,5]
y1 = [3,7,1,9,5]

x2 = [3,4,6,1,5]
y2 = [1,7,2,8,9]

pyplot.plot(x1,y1,"o:r")
pyplot.plot(x2,y2,"*:g")
```

```python
height = [150,155,145,153,140]
weight = [80,67,77,82,98]
pyplot.plot(height,weight,"*:r")
pyplot.xlabel("Height in cm")
pyplot.ylabel("Weight in kg")#these three methods helps to understand the graph for users
pyplot.title("Height Weight graph")
pyplot.grid()
pyplot.show()
```

```python
pyplot.bar(height,weight)
```
