```python
Gini
---------
In a random forest, Gini is a metric called the Gini impurity or Gini index that is used to determine the quality of a split in a decision tree.
 It measures the diversity of the classes within a node, and the goal is to minimize this impurity as you move down the tree, making the splits more pure and accurate.
  A Gini score of 0 represents a perfectly pure node (all data points belong to one class), 
  while a score of 0.5 represents the maximum impurity for a binary split (an even mix of classes). 

Equation to find the gini----
1-E(pi^2)


-Minimizing impurity: A split is considered good if the Gini impurity of the resulting
 child nodes is lower than the impurity of the parent node. The algorithm chooses 
 the split that results in the greatest reduction in Gini impurity.
 
-Gini importance: In a random forest, Gini importance is an extension
 that measures how much a feature contributes to the overall accuracy.
  It's calculated based on the total reduction in Gini impurity
   that a feature provides across all trees in the forest
```
