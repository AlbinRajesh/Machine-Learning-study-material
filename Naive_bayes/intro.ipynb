```python
Naive bayes
=========================
Probability
------------------------------
chances of occuring an event
p(A)

THREE TYPES OF PROBABILITY==================================

marginal probability
-----------------------------
chance for occuring a single event

joint probability
------------------
chance for occruing two events together p(A|b)

conditional probability
------------------------------
The probability of an event happening, given that another event has already happened It's written as (p(A|B))
which means "the probability of A given B".
```

```python
weather           temperature           play

sunny              hot                  no

sunny              mild                 no

rainy              mild                 yes

rainy              cool                 yes

sunny              cool                 no


no we need to check if  wether  = rainy and temperature is hot (event B already occured) will they play (event A)
```

```python
Prior Probability
_________________
probability of event A(will they play) without seeing any evidence (already occured event (wether and temp))
p(play = yes) = (total no of yes)/(total no of outcomes) = 2/5 >> 0.4
p(play = no) = (total no of no )/(total no of outcomes) = 3/5
```

```python
Likelyhood
--------------

wether  = rainy and temperature

p(wether = rainy | play = yes) = 2/2 = 1
p(wether = rainy | play = no) = 0/2 = 0

p(temp = hot | play = yes ) = 0/1 = 0
p(temp = hot | play = no ) = 1/1 = 1
```

```python

p(yes|Rainy,Hot) = p(yes) * p(rainy|yes) * p(hot|yes)   >>> 2/5 * 1 * 0 >> 0
p(no|rainy,hot) = p(no) * p(rainy|no) * p(hot|no)   >>> 3/5 * 0 * 1 >> 0
```

```python
3)   income        credit_score       loan approved

    high           good               yes

    high           bad                no

    medium         good               yes

    low            good               no

    low             bad               no

    medium          bad                yes

    we need to find if the income = high and credit score =  good ,will the loan be approved
```

```python
Prior Probability
---------------------
p(loan = yes) = (total no of yes)/(total no of outcomes) = 3/6 = 0.5
p(loan = no) = (total no of yes)/(total no of outcomes) = 3/6 = 0.5

Likelyhood
--------------
p(income = high | loan = yes) = 1/3 = 0.3
p(income = high | loan = no) = 1/3 = 0.3

p(credit_score = good | loan = yes) = 2/3 = 0.6
p(credit_score = good | loan = no) = 1/3 = 0.3

Applying bayes theorem
--------------------------
p(yes|income,credit_score) =  p(yes) * p(income|yes) * p(credit_score|yes) = 0.5 * 0.3 * 0.6 =
 0.9

p(no|income,credit_score) =  p(no) * p(income|no) * p(credit_score|no) = 0.5 * 0.3 * 0.3 =
 0.04

the chances of loan approving when the income = high and credit score = good  is 0.6 and not approving is 0.04

the loan will be approved--------
```
