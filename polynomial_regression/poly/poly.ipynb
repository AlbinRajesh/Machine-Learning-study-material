```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("poly.csv")
df
```

```python
df.drop(labels=["sno"],axis =1,inplace=True)
```

```python
df
```

```python
x = df.iloc[:,:-1]
x
y = df.iloc[:,-1]
y
```

```python
from sklearn.linear_model import LinearRegression
le = LinearRegression()
le.fit(x,y)
```

```python
y_pred = le.predict(x)
```

```python
print(y_pred)
print(y)
```

```python
import matplotlib.pyplot as plt 
```

```python
from sklearn.preprocessing import PolynomialFeatures
features = PolynomialFeatures(degree=2)
x_poly = features.fit_transform(x)  #
x_poly
model1 = LinearRegression()
model1.fit(x_poly,y)
```

```python
y_poly = model1.predict(x_poly)
y_poly
```

```python
plt.scatter(x,y,color="red")
plt.plot(x,y_pred)
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
print(mean_absolute_percentage_error(y,y_pred))
print(r2_score(y,y_pred))
```

```python
print(mean_absolute_percentage_error(y,y_poly))
print(r2_score(y,y_poly))
```
