```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('hypothyroid.csv')
df
```

```python
df.drop(labels="age", axis=1, inplace=True)
```

```python
df.isna().sum()
```

```python
df["binaryClass"].unique()
```

```python
df.replace("?", np.nan, inplace=True)
```

```python
df.isna().sum()
```

```python
df.drop(labels=['TBG'], axis=1, inplace=True    )
```

```python
df.isna().sum()
```

```python
df.isna().sum()
```

```python
df.dropna(subset=["sex"],inplace=True)
```

```python
df.isna().sum()
```

```python
nul_value = ["TSH","T3","TT4","T4U","FTI"]
df[nul_value].dtypes
```

```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy='mean')
df[nul_value] = si.fit_transform(df[nul_value])
```

```python
df.isna().sum()
```

```python
char_col = df.select_dtypes(include=['object']).columns
df[char_col]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in char_col:
    df[col] = le.fit_transform(df[col])
```

```python
df
```

```python
x = df.iloc[:,:-1]
x
y = df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()  
dtc.fit(x_train, y_train)
y_pred_dtc = dtc.predict(x_test)
```

```python
print(accuracy_score(y_test, y_pred_dtc))
```
