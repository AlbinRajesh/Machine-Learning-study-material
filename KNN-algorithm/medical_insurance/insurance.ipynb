```python
import pandas as pd
import numpy as np
# 0.8714333333333333
```

```python
df = pd.read_csv("medical_insurance.csv")
df
```

```python
df.isna().sum()
```

```python
df.drop(labels=["person_id"],axis=1,inplace=True)
df.drop(labels=["region"], axis=1, inplace=True)
df
```

```python
df
```

```python
x = df.drop(labels="is_high_risk",axis=1)
x
y = df["is_high_risk"]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.2,random_state=42)
```

```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy="most_frequent")
x_train[["alcohol_freq"]] = si.fit_transform(x_train[["alcohol_freq"]])
x_test[["alcohol_freq"]] = si.fit_transform(x_test[["alcohol_freq"]])
```

```python
x_test.isna().sum()
```

```python
x_train.isna().sum()
```

```python
char_col = df.select_dtypes(include="object").columns
char_col
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore',sparse_output=False)
encoder.fit(x_train[char_col])
x_train = encoder.transform(x_train[char_col])
x_test = encoder.transform(x_test[char_col])
```

```python
x_train
```

```python
# df2 = df.copy()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
y_pred_naive = gnb.predict(x_test)
```

```python
print(accuracy_score(y_test, y_pred_naive))
```
