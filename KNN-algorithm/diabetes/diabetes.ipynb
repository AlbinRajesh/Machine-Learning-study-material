```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("diabetes.csv")
df.isna().sum()
```

```python
#target = Outcome
#no missing values
df
```

```python
X = df.iloc[:,:-1]
X
y = df.iloc[:,-1]
y
```

```python
df.dtypes
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
#here firstly the data is being divided as training data and testing data for better data confidentiality
```

```python
# from sklearn.preprocessing import MinMaxScaler = it is one of the scalling method
from sklearn.preprocessing import StandardScaler#this is another method used for scaling
scaler = StandardScaler()
#only training data should be fit and transformed 
#if we fit the test data it may lead to data leakage and would affect the model so dont give it like that
X_train = scaler.fit_transform(X_train)
x_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
y_pred = knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,y_test))
```

```python
from sklearn.metrics import confusion_matrix
cm  = confusion_matrix(y_test,y_pred)
print(cm)
#[[70 17]
# [28 39]] == this is a confusion matrix [[ TP FP
                               #           FN TN ]] = [[ TP = 70  FP = 17
                               #                         FN = 28  TN = 39 ]]
                                                     
 
```
