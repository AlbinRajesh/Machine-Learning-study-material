```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Student_performance_data _.csv")
df
```

```python
df.isna().sum()#first step is to check null values
```

```python
#secound step is to check data types and convert then into same data type
df[["StudyTimeWeekly","GPA","GradeClass"]]= df[["StudyTimeWeekly","GPA","GradeClass"]].astype(int)
df.dtypes
```

```python
df
```

```python
df.drop(labels=["StudentID","Age","Gender"], axis=1, inplace=True)
#third step is to drop unnecesary columns
```

```python
df
```

```python
#identifying the fetures and target columns
# X = fetures
# y = target
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
y
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
#creating an object for the class MIN max scaler
scaler.fit(x)#fetures values simlier range 0-1 #chil case il values thamil ul difference veluthayirikum eg(2,40,4,60,120),
#ithupole ula data vech cheythal acuraccy valare kurav ayirikum ,
#so aa prshnm solve cheyan min max scaler use cheyyum ,ithupole kure methods und ath solve cheyan ivide use cheyunath minmax scaler aanu\
#it orginal data kk currespondent ayittula 0-1 range il data convert cheyyum
x_scaled = scaler.transform(x)
x_scaled
#these are the steps of feature scaling
#it is the first step in ML model building after preparing the data
```

```python
# dividing the data into training and testing data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled,y,test_size=0.2) 
# ------------------------------------------------------------  
#x_train = 80% training data(features)
#x_test = 20% testing data(features)
#y_train = 80% training data(target)
#y_test = 20% testing data(target)
# ------------------------------------------------------------
#size tells the how it should take for the testing purpose,
# Here it takes 20% testing (which is we specified like (0.2),(0,3)means 30% for testing purpose (0.5) means 50% for testing purpose) and 80% for training purpose
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
#here the model default value for n_neighbors is 5(this is the odd number we take)
knn.fit(x_train,y_train)
```

```python
y_pred = knn.predict(x_test)
y_pred
```

```python
y_test.values
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
#0.6617954070981211 :: 66% is the accurace of this model
#it changes every time because of the random state
```
