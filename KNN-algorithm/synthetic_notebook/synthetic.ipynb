```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("synthetic_dataset.csv")
df
```

```python
df = df.dropna(subset=["Category"])#this remmoves the rows that are null from the category coloumn
```

```python
df.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
numerical_col =["Price","Rating","Discount"]
num_imputer = SimpleImputer(strategy="median")
df[numerical_col] = num_imputer.fit_transform(df[numerical_col])
```

```python
stock_imputer = SimpleImputer(strategy="most_frequent")
df[["Stock"]] = stock_imputer.fit_transform(df[["Stock"]])
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Category"]=le.fit_transform(df["Category"])
df["Stock"]=le.fit_transform(df["Stock"])
```

```python
df
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[["Price","Rating","Discount"]] = scaler.fit_transform(df[["Price","Rating","Discount"]])
```

```python
df
```

```python
x = df.iloc[:,1:]
x
y = df.iloc[:,0]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn  = KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
y_pred  = knn.predict(x_test)
y_pred
```

```python
y_test.values
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```
